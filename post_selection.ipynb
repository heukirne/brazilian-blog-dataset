```python
import pandas as pd
import numpy as np
posts = pd.read_csv('posts_count_me_pol.csv.gz', compression='gzip', error_bad_lines=False)
```

```python
posts = posts[posts['me'] > 2]
posts = posts[posts['polarity'] > 2]
len(posts)
```

```python
sample = posts.sample(3000) 
sample.to_csv('posts_sample.csv')
sample
```

```python
blogs = sample[['posts.csv']].groupby(['posts.csv'])
autores = sample[['authorID']].groupby(['authorID'])
print('Postagens: %i' %len(sample))
print('Blogs: %i' %len(blogs))
print('Autores: %i' %len(autores))
```
