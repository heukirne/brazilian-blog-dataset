```python
import pandas as pd
import numpy as np
from liwc import LIWC
liwc_class = LIWC()
from naivesumm import NaiveSummarizer
nv = NaiveSummarizer()
posts = pd.read_csv('posts_count_me_pol.csv.gz', compression='gzip', error_bad_lines=False)
```

```python
posts = posts[posts['me'] > 2]
posts = posts[posts['polarity'] > 2]
len(posts)
```

```python
def cutContent(x): 
    try:
        return nv.summarize(str(x['content']),5)
    except:
        return 0
    
def buildMe(x): 
    try:
        texts = []
        texts.append(x['cutContent'])
        mtx = liwc_class.build_features(texts)
        return mtx[0][3]
    except:
        return 0
    
def buildPolarity(x): 
    try:
        texts = []
        texts.append(x['cutContent'])
        mtx = liwc_class.build_features(texts)
        return mtx[0][27] + mtx[0][28]
    except:
        return 0
```

```python
sample = posts.sample(3000) 
sample['cutContent'] = sample.apply(cutContent,axis=1)
sample['cutme'] = sample.apply(buildMe,axis=1)
sample['cutpolarity'] = sample.apply(buildPolarity,axis=1)
```

```python
sample = sample[sample['cutme'] > 2]
sample = sample[sample['cutpolarity'] > 2]
len(sample)
```

```python
resample = sample.sample(1000)
resample.to_csv('posts_sample.csv')
resample
```

```python
blogs = resample[['posts.csv']].groupby(['posts.csv'])
autores = resample[['authorID']].groupby(['authorID'])
print('Postagens: %i' %len(resample))
print('Blogs: %i' %len(blogs))
print('Autores: %i' %len(autores))
```
