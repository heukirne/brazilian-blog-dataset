```python
import pandas as pd
import numpy as np
posts = pd.read_csv('posts_count_me.csv.gz', compression='gzip', error_bad_lines=False)
```

```python
%matplotlib inline
iamHist = posts['me']
iamHist = iamHist[iamHist > 0]
iamHist = iamHist[iamHist < 100]
iamHist.plot.hist(bins=40)
```
