```python
import pandas as pd
from liwc import LIWC
liwc_class = LIWC()

portuguese_stories = pd.read_csv('posts_score.csv.gz', error_bad_lines=False)
#portuguese_stories = portuguese_stories.sample(200)
portuguese_stories.shape
```

```python
mtx = liwc_class.build_features(portuguese_stories.content)
mtx.shape
```

```python
import re

def wc(x): 
    try:
        return len(re.findall(r'\w+', x['content']))
    except:
        return 0
    
def wps(x): 
    try:
        return x['wc'] / x['content'].count(".")
    except:
        return x['wc']
```

```python
columns = ['funct','pronoun','ppron','i','we','you','shehe','they','ipron','article','verb','auxverb','past','present','future','adverb','preps','conj','negate','quant','number','swear','social','family','friend','humans','affect','posemo','negemo','anx','anger','sad','cogmech','insight','cause','discrep','tentat','certain','inhib','incl','excl','percept','see','hear','feel','bio','body','health','sexual','ingest','relativ','motion','space','time','work','achieve','leisure','home','money','relig','death','assent','nonfl','filler']
df_liwc = pd.DataFrame(mtx, columns=columns)
```

```python
df_liwc['score'] = portuguese_stories.score.values
portuguese_stories['wc'] = portuguese_stories.apply(wc,axis=1)
portuguese_stories['wps'] = portuguese_stories.apply(wps,axis=1)
df_liwc['wc'] = portuguese_stories['wc'].values
df_liwc['wps'] = portuguese_stories['wps'].values
```

```python
df_liwc.to_csv('portuguese_stories_liwc.csv.gz', compression='gzip')
df_liwc.shape
```
