```python
import pandas as pd
from nltk.tokenize import word_tokenize
from liwc import LIWC
import LexRank
liwc_class = LIWC()
sample = pd.read_csv('posts_sample.csv')
```

```python
def firstPart(x): 
    try:
        return ' '.join(word_tokenize(x['content'])[:100])
    except:
        return 0
    
def readMore(x): 
    try:
        return ' '.join(word_tokenize(x['content'])[100:])
    except:
        return 0
```

```python
sample['firstPart'] = sample.apply(firstPart,axis=1)
sample['readMore'] = sample.apply(readMore,axis=1)
```

```python
resample = sample.sample(1000)
resample = resample.drop(resample.columns[[0,1,2,3,4]], 1)
resample = resample.rename(columns={'posts.csv': 'blogID'})
resample.to_csv('posts_resample.csv')
resample
```

```python
blogs = resample[['blogID']].groupby(['blogID'])
autores = resample[['authorID']].groupby(['authorID'])
print('Postagens: %i' %len(resample))
print('Blogs: %i' %len(blogs))
print('Autores: %i' %len(autores))
```
